```
df = pd.read_json(path_or_buf = 'first500final.json')
```

```
df = df.T
```

```
df["RATING"] = pd.to_numeric(df["RATING"])
```

```
df.head()
```

```
datos_modelo = [0, 0]
datos_diseno = [0, 0]
datos_composicion = [0, 0]
datos_otro = [0, 0]
```

```
datos_modelo = [df[df["LABEL"]=="Modelo"].sum()["RATING"], df[df["LABEL"]=="Modelo"].count()["RATING"]]
datos_diseno = [df[df["LABEL"]=="Diseo"].sum()["RATING"], df[df["LABEL"]=="Diseo"].count()["RATING"]]
datos_composicion = [df[df["LABEL"]=="Composicin"].sum()["RATING"], df[df["LABEL"]=="Composicin"].count()["RATING"]]
datos_otro = [df[df["LABEL"]=="Otro"].sum()["RATING"], df[df["LABEL"]=="Otro"].count()["RATING"]]
```

```
print(datos_modelo, datos_diseno, datos_composicion, datos_otro)
```

```
labels = ["modelo", "diseno", "composicion", "otro"]
avgs = [datos_modelo[0]/datos_modelo[1], datos_diseno[0]/datos_diseno[1], datos_composicion[0]/datos_composicion[1], datos_otro[0]/datos_otro[1]]
```

```
avgs
```

```
plt.bar(labels, avgs)
plt.plot()
```
